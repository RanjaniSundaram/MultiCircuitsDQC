```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tabulated2.csv')
df
```

```python
df['%gain']=(df['MartinezCost']-df['OurCost'])/df['MartinezCost']*100
```

```python
df['FracCZ']=df['numCZ']/df['numGates']
df['1+epsilon']=df['size']*df['numPartitions']/df['numQubits']
```

```python
df.describe()
```

```python
df['FracCZbuckets']=pd.cut(df['FracCZ'], np.arange(0, 1+0.3, 0.3))


df['Qubitbuckets']=pd.cut(df['numQubits'], np.arange(0, 125, 25))

df['partitionbuckets']=pd.cut(df['numPartitions'], np.arange(0, 30, 7))

df['epsilonbuckets']=pd.cut(df['1+epsilon'], np.arange(1, 3, 0.3))
df
```

```python
VaryQubits=pd.read_csv('final_tabulated_Qubits2.csv')
VaryQubits

VaryQubits['Omgain']=(VaryQubits['MartinezCost']-VaryQubits['OM_cost'])/(VaryQubits['MartinezCost'])*100
VaryQubits['Gstar_simplegain']=(VaryQubits['MartinezCost']-VaryQubits['Gstar_simple_cost'])/(VaryQubits['MartinezCost'])*100
VaryQubits['Gstar_approxgain']=(VaryQubits['MartinezCost']-VaryQubits['Gstar_approx_cost'])/(VaryQubits['MartinezCost'])*100
VaryQubits['GstarLPgain']=(VaryQubits['MartinezCost']-VaryQubits['GstarLP_cost'])/(VaryQubits['MartinezCost'])*100


VaryQubits2=VaryQubits.groupby(['numQubits']).mean().reset_index()

VaryQubits2 = VaryQubits2[VaryQubits2.numQubits % 10 ==0]

ind=VaryQubits2['numQubits'].values

MartinezCost=[100]*len(ind)
OM=100-VaryQubits2['Omgain']
G_star_simple=100-VaryQubits2['Gstar_simplegain']
G_star_approx=100-VaryQubits2['Gstar_approxgain']
G_starLP=100-VaryQubits2['GstarLPgain']

width=2
width2=1
fig = plt.figure(figsize=(16, 6))
ax = fig.add_subplot()
rects1 = ax.bar(ind, MartinezCost, width2, color='royalblue', label='Martinez-19')

rects2 = ax.bar(ind+width, OM, width2, color='seagreen', label='Home-Cover' )
rects3 = ax.bar(ind+width+width, G_star_simple, width2, color='orange', label= 'G*_simple' )
rects4 = ax.bar(ind+width+width+width, G_star_approx, width2, color='yellow', label= 'G*_approx' )
rects5 = ax.bar(ind+width+width+width+width, G_starLP, width2, color='red', label= 'G*_LP' )
#rects6 = ax.bar(ind+width+width+width+width+width/, MartinezCost, width2, color='white' )

ax.set_xticks(ind + width+width )
ax.set_xticklabels(ind)
ax.set_ylabel('Percentage of ebits used')
ax.set_xlabel('Number of qubits')
plt.title("Percentage of ebits gained or lost compared to Martinez")
plt.legend()
plt.savefig("VaryQubits.png", pad_inches=0.5)
plt.show()
```
